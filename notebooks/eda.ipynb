```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv('../data/raw/MachineLearningRating_v3.txt', sep='|')
df.head()
```

```python
df.info()
df.describe(include='all')
df.shape
```

```python
missing = df.isnull().mean().sort_values(ascending=False)
missing.head(20)
```

```python
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'], errors='coerce')
```

```python
df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
df['LossRatio'].describe()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Province', y='LossRatio', estimator=np.mean)
plt.title("Average Loss Ratio by Province")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../visuals/loss_ratio_province.png')
plt.show()
```

```python
monthly = df.groupby(df['TransactionMonth'].dt.to_period('M')).TotalClaims.sum()

monthly.plot(figsize=(12,5))
plt.title("Monthly Total Claims Trend")
plt.xlabel("Month")
plt.ylabel("Claims")
plt.tight_layout()
plt.savefig('../visuals/monthly_claims_trend.png')
```

```python
severity = df[df['TotalClaims'] > 0].groupby('make').TotalClaims.mean().sort_values(ascending=False).head(15)

plt.figure(figsize=(10,6))
severity.plot(kind='barh')
plt.title("Top 15 Vehicle Makes by Claim Severity")
plt.tight_layout()
plt.savefig('../visuals/claim_severity_by_make.png')
plt.show()
```
